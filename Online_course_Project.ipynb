We notice a lot of formats in the rows. We need this in a single unit: Hour.

```python
import re

def clean_duration(duration):
    if pd.isna(duration):
        return None

    # Normalization
    duration = duration.lower()
    duration = duration.replace('horas', 'hour').replace('meses', 'month').replace('semanas', 'week')
    duration = re.sub(r'[\n]', '', duration)
    duration = re.sub(r'\s+', ' ', duration).strip()
    duration = duration.replace('approx.', '').replace('approximately', '').replace('estimated', '').strip()

    # Pattern 1: "X months at Y hours a week"
    match = re.search(r'(\d+)\s*months?\s*at\s*(\d+)\s*hours?\s*a\s*week', duration)
    if match:
        months = int(match.group(1))
        hours_per_week = int(match.group(2))
        return months * 4 * hours_per_week

    # Pattern 2: "X - Y Weeks"
    match = re.search(r'(\d+)\s*-\s*(\d+)\s*weeks?', duration)
    if match:
        weeks_start = int(match.group(1))
        weeks_end = int(match.group(2))
        return ((weeks_start + weeks_end) / 2) * 5 # Assume 5 hours per week

    # Pattern 3: Simple Units ("X months", "X weeks", "X hours", "X minutes")
    match_month = re.search(r'(\d+)\s*months?', duration)
    if match_month:
        months = int(match_month.group(1))
        return months * 4 * 5 # Assume 4 weeks per month and 5 hours per week

    match_week = re.search(r'(\d+)\s*weeks?', duration)
    if match_week:
        weeks = int(match_week.group(1))
        return weeks * 5 # Assume 5 hours per week

    match_hour = re.search(r'(\d+)\s*hours?', duration)
    if match_hour:
        hours = int(match_hour.group(1))
        return hours

    match_minute = re.search(r'(\d+)\s*minutes?', duration)
    if match_minute:
        minutes = int(match_minute.group(1))
        return minutes / 60

    # Handle cases like "X Hour"
    match_single_hour = re.search(r'(\d+)\s*hour', duration)
    if match_single_hour:
        hours = int(match_single_hour.group(1))
        return hours

    return None

# Apply the cleaning function to create the new 'Duration_Hours' column
df['Duration_Hours'] = df['Duration'].apply(clean_duration)

# Display the first few rows with the new column and check info
print("DataFrame with new 'Duration_Hours' column:")
display(df[['Duration', 'Duration_Hours']].head())

print("\nInfo of the DataFrame with 'Duration_Hours':")
df.info()
```

#Master Ctageory

```python
df['Master_Category'].unique()
```

We clean the categories by mapping tables.

One map cleans the categories we already have. and the second maps key words in title to correct categories

```python
translation_map = {
    # Spanish
    'ciencia de datos': 'Data Science',
    'negocios': 'Business',
    'ciencias de la computación': 'Computer Science',

    # Portuguese
    'negócios': 'Business',
    'tecnologia da informação': 'Information Technology',

    # Japanese
    'データサイエンス': 'Data Science',

    # Chinese
    '计算机科学': 'Computer Science'
}

```

```python
df.groupby('Master_Category').head(3)
```

```python
df[df['Master_Category'].isnull()]
```

```python
keyword_map = {
    'machine learning': 'Data Science',
    'data science': 'Data Science',
    'deep learning': 'Data Science',
    'artificial intelligence': 'Data Science',
    'analytics': 'Data Science',
    'sql': 'Data Science',
    'tableau': 'Data Science',
    'data': 'Data Science',
    'analyst': 'Data Science',

    # Computer Science
    'python': 'Computer Science',
    'java': 'Computer Science',
    'c++': 'Computer Science',
    'javascript': 'Computer Science',
    'computer science': 'Computer Science',
    'software engineering': 'Computer Science',
    'web development': 'Computer Science',
    'front-end': 'Computer Science',
    'react': 'Computer Science',

    # Business
    'project management': 'Business',
    'business': 'Business',
    'finance': 'Business',
    'accounting': 'Business',
    'marketing': 'Business',
    'excel': 'Business',

    # Information Technology
    'google cloud': 'Information Technology',
    'aws': 'Information Technology',
    'azure': 'Information Technology',
    'cybersecurity': 'Information Technology',
    'it support': 'Information Technology',
    'scrum': 'Information Technology',
    'data engineering': 'Information Technology',
    'engineer':'Physical Science And Engineering', #general engineer

    # Personal Development
    'leadership': 'Personal Development',
    'career': 'Personal Development',
    'communication': 'Personal Development',

    # Language
    'english': 'Language Learning',
    'spanish': 'Language Learning',
    'speak': 'Language Learning',

    # Other
    'psychology': 'Social Sciences',
    'music': 'Arts and Humanities',
    'health': 'Health'
}
```

```python
# Step 1: Standardize the 'Master_Category' column to lowercase
df['Master_Category'] = df['Master_Category'].astype(str).str.lower()
```

```python
# Step 2: Translate the languages
# Use .replace() instead of .map().
# .replace() changes what's in the map and *leaves everything else alone*.
df['Master_Category'] = df['Master_Category'].replace(translation_map)
```

```python
#Step 3: Capitalize the good data and fix 'nan' / 'other' strings
# We now have 'data science', 'business', 'nan', 'other' etc.
df['Master_Category'] = df['Master_Category'].str.title()
# Turn the string 'Nan' and 'Other' back into real nulls so we can fill them
df['Master_Category'] = df['Master_Category'].replace(['Nan', 'Other'], np.nan)
```

```python
# Step 4: Fill the *real* nulls using the 'Title' column
# We loop through the keyword_map and fill the nulls.
# This is the correct way to do the keyword search.
for keyword, category in keyword_map.items():

    # Find rows where Category is still null AND Title contains the keyword
    mask = (
        df['Master_Category'].isnull() &
        df['Title'].str.contains(keyword, case=False, na=False)
    )

    # Fill those rows with the correct category
    df.loc[mask, 'Master_Category'] = category

# Step 5: Fill any remaining nulls with 'Other'
df['Master_Category'] = df['Master_Category'].fillna('Other')
```
